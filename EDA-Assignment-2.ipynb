```python
# importing library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Dataset/amazon.csv")
```

```python
df.info()
```

```python
df.describe()
```

```python
average_rating_per_category = df.groupby('category')['rating'].mean()
print(average_rating_per_category)
```

```python
df['discount_percentage'] = ((df['actual_price'] - df['discounted_price']) / df['actual_price']) * 100
average_discount_percentage_per_category = df.groupby('category')['discount_percentage'].mean()
average_discount_percentage_per_category
```

```python
most_popular_product_names = df['product_name'].value_counts().nlargest(5)
most_popular_product_names
```

```python
# Assuming 'product_name' contains keywords
most_popular_keywords = df['product_name'].str.split().explode().value_counts().nlargest(5)
most_popular_keywords
```

```python
most_popular_reviews = df['review_content'].value_counts().nlargest(5)
most_popular_reviews
```

```python
correlation_discounted_price_rating = df['discounted_price'].corr(df['rating'])
correlation_discounted_price_rating
```

```python
top_5_categories_highest_ratings = df.groupby('category')['rating'].mean().nlargest(5)
top_5_categories_highest_ratings
```

```python
# Convert 'rating_count' column to numeric type
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

# Remove NaN values after conversion
df = df.dropna(subset=['rating_count'])

# Now calculate the quantile
rating_count_outliers = df[df['rating_count'] > df['rating_count'].quantile(0.95)]
rating_count_outliers
```

# Spotify Dataset


```python
# Load the dataframe
df = pd.read_csv("Dataset/spotify.csv")

# Check for missing values
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)

# Remove duplicate rows
df.drop_duplicates(inplace=True)
```

```python
# Visualize distribution of popularity
plt.hist(df['Popularity'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Track Popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot of popularity vs duration
plt.scatter(df['Duration (ms)'], df['Popularity'], alpha=0.5)
plt.title('Popularity vs Duration of Tracks')
plt.xlabel('Duration (ms)')
plt.ylabel('Popularity')
plt.show()
```

```python
# Countplot of tracks per artist
plt.figure(figsize=(10, 6))
sns.countplot(y='Artist', data=df, order=df['Artist'].value_counts().index)
plt.title('Number of Tracks per Artist')
plt.xlabel('Number of Tracks')
plt.ylabel('Artist')
plt.show()
```

```python
# Top 5 least popular tracks
least_popular_tracks = df.nsmallest(5, 'Popularity')[['Artist', 'Track Name', 'Popularity']]
print("Top 5 Least Popular Tracks:\n", least_popular_tracks)
```

```python
# Calculate average popularity for each artist
average_popularity_per_artist = df.groupby('Artist')['Popularity'].mean().sort_values(ascending=False).head(5)
print("Average Popularity per Artist (Top 5):\n", average_popularity_per_artist)
```

```python
# List most popular tracks for each of the top 5 most popular artists
top_5_most_popular_artists = df.groupby('Artist')['Popularity'].mean().nlargest(5).index
for artist in top_5_most_popular_artists:
    most_popular_track = df[df['Artist'] == artist].nlargest(1, 'Popularity')['Track Name'].values[0]
    print(f"Most Popular Track for {artist}: {most_popular_track}")
```

```python
# Pair plot of numerical variables
sns.pairplot(df)
plt.show()
```

```python
# Box plot of duration across different artists
plt.figure(figsize=(10, 6))
sns.boxplot(x='Artist', y='Duration (ms)', data=df)
plt.xticks(rotation=45)
plt.title('Duration of Tracks Across Different Artists')
plt.xlabel('Artist')
plt.ylabel('Duration (ms)')
plt.show()
```

```python
# Violin plot of track popularity across different artists
plt.figure(figsize=(18, 8))
sns.violinplot(x='Artist', y='Popularity', data=df)
plt.xticks(rotation=45)
plt.title('Distribution of Track Popularity Across Different Artists')
plt.xlabel('Artist')
plt.ylabel('Popularity')
plt.show()
```
